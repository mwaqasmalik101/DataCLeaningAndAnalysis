<h3> <center> Importing Libraries </center> </h1>

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.linspace(5,100, 100)
y = np.linspace(10, 1000, 100)
```

```python
plt.plot(x,y)
```

<h3> <center> Line Graph </center> </h1>

```python
plt.plot([1,2,3,4,5],[5,10,15,20,25])
```

<center > Labeling Axis and Title </center>

```python
plt.grid()
plt.plot(x,y)
plt.title('OHm\'s Law')
plt.xlabel('Current')
plt.ylabel('Voltage')

plt.xlim([20, 80])
plt.ylim([200, 800])
plt.show()
```

<h3> <center> Discrete Graph </center> </h1>

```python
x = np.linspace(1,10, 20)
y = x*2
plt.plot(x,y, 'b+')
plt.show()
```

<h3> <center> Ploting different lines on one graph</center></h3>

```python
x = np.linspace(0,5,10)
y = np.linspace(3,6,10)
plt.plot(x,y, 'r-', x,y**2, 'b+', x,y**3, 'g^')
plt.show()
```

<h3> <center>Ploting Subgraph </center> </h3>

```python
x = np.linspace(1,10, 100)
y = np.log(x)

plt.figure(1)

plt.subplot(121)
plt.title("y = log(x)")
plt.plot(x, y)

plt.subplot(122)
plt.title("y = log(x)**2")
plt.plot(x,y**2)
plt.show()
```

```python
x = np.linspace(1,10, 100)

plt.figure(1)

# subplot 1
plt.subplot(2,2, 1)
plt.title("Linear Graph ")
plt.plot(x, x)

# subplot 1
plt.subplot(2, 2 , 2)
plt.title("Linear Graph ")
plt.plot(x, x*x)


plt.figure(2)
# subplot 3
plt.subplot(2,2,1)
plt.title("Cubic Graph ")
plt.plot(x, x**3)

# subplot 4
plt.subplot(2,2,2)
plt.title("Exponential ")
plt.plot(x, x**x)
```
